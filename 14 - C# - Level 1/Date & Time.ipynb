# Working with Date and Time in C#


```
DateTime Dt1 = new DateTime(); // DataTime Is struct ,assigns default value 01/01/0001 00:00:00

// will print the default date and time
Console.WriteLine("The Default Date & Time Is: {0}",Dt1); 
```

```
//assigns year, month, day, hour, min, seconds, UTC timezone

DateTime UTC = new DateTime(2026, 8, 3, 5, 10, 23,DateTimeKind.Utc);

Console.WriteLine(UTC);
```

```
DateTime Now = DateTime.Now; // to get the current date and time
Console.WriteLine(Now);
```

# Ticks

```
long Ticks = DateTime.Now.Ticks; // long number represents the time that have passed since 1/1/1 12:00 am untill now 
Console.WriteLine("The Number Of Ticks in The current TimeDate Object: {0}",Ticks);
long Ticks2 = DateTime.Now.AddSeconds(10).Ticks;

Console.WriteLine("The Diffrence between 2 times is: {0}",(Ticks2-Ticks));


```

```
Console.WriteLine(DateTime.MinValue.Ticks);//min value of ticks
Console.WriteLine(DateTime.MaxValue.Ticks);// max value of ticks
```

# Datetime Static Fields

```
DateTime dt = DateTime.Now;  //returns current date and time

DateTime Today = DateTime.Today; // returns today's date

DateTime UTC = DateTime.UtcNow;// returns today's date

DateTime Max = DateTime.MaxValue;

DateTime Min = DateTime.MinValue;

Console.WriteLine(dt);
Console.WriteLine(Today);
Console.WriteLine(UTC);
Console.WriteLine(Max);
Console.WriteLine(Min); // = ticks start point
```
