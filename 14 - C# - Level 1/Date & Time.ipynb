# Ticks

```
long Ticks = DateTime.Now.Ticks; // long number represents the time that have passed since 1/1/1 12:00 am untill now 
Console.WriteLine("The Number Of Ticks in The current TimeDate Object: {0}",Ticks);
long Ticks2 = DateTime.Now.AddSeconds(10).Ticks;

Console.WriteLine("The Diffrence between 2 times is: {0}",(Ticks2-Ticks));


```

```
Console.WriteLine(DateTime.MinValue.Ticks);//min value of ticks
Console.WriteLine(DateTime.MaxValue.Ticks);// max value of ticks
```

# Time Span

|              | DateTime    | TimeSpan  |
| ------------ | ----------- | --------- |
| يمثل         | لحظة        | مدة       |
| مرتبط بتاريخ | نعم         | لا        |
| مثال         | 8 مارس 2026 | 3 ساعات   |
| يستخدم في    | timestamps  | durations |


```
DateTime dt = DateTime.Now; // the current date and time 
DateTime dt2 = new DateTime(2006, 8, 1,1,12,33); // Specific Moment Of Time
// (NOT HOW LONG THE TIME)
Console.WriteLine(dt2.Year);
Console.WriteLine(dt2.Month);

TimeSpan s = new TimeSpan(1, 20, 33); // it isn't specific momnent of time , (HOW LONG THE TIME)
Console.WriteLine("_____________");
Console.WriteLine(s.Hours);
Console.WriteLine(s.Minutes);
Console.WriteLine(s.Seconds);
```

```
//   Subtraction of two dates results in TimeSpan /// 
DateTime d = DateTime.Now;
TimeSpan diff =  DateTime.Now.AddHours(1)-d; // the difference between 2 different times

Console.WriteLine(diff);

// another way 
DateTime NOW = DateTime.Now;
DateTime After1Year = new DateTime(2027, 1, 1, 12, 00, 00);

// الطرح بين تاريخين مختلفين بشكل او باخر بينتج عنه فجوه زمنيه يعني فرق بين زمنين مختلفين 
TimeSpan Diff = After1Year.Subtract(NOW);
Diff = After1Year - NOW; // another way it is correct also 
Console.WriteLine($"The DiFF Between Now And 2027 is: {Diff.Days}");
```

```
DateTime d = DateTime.Now; // the current date  and time 

TimeSpan PersiodToAdd = new(1, 1, 1); // 1h,1m,1s
DateTime s = d.Add(PersiodToAdd); // add the period to current time 

Console.WriteLine("The Time After Adding Specific Period");
Console.WriteLine($"The Hour:{s.Hour}");
Console.WriteLine($"The Min:{s.Minute}");
Console.WriteLine($"The Second:{s.Second}");
Console.WriteLine($"The Total Hours to add is: {PersiodToAdd.TotalHours}");

Console.WriteLine($"The Total Mins to add is: {PersiodToAdd.TotalMinutes}");

Console.WriteLine($"The Total Seconds to add is: {PersiodToAdd.TotalSeconds}");
```

```

TimeSpan H = TimeSpan.FromHours(3);
TimeSpan M =TimeSpan.FromMinutes(1);
TimeSpan S =TimeSpan.FromSeconds(100);
```

###### DateTime + TimeSpan = DateTime
###### DateTime - DateTime = TimeSpan

```
DateTime NOW = DateTime.Now;
DateTime After3Hours = NOW.Add(H);
Console.WriteLine("After Adding 3 Hours: {0}", After3Hours);

DateTime AfterAdding1M = NOW.Add(M);
Console.WriteLine("The Time After Adding 1M : {0}", AfterAdding1M);

DateTime AfterAdd100Secs = NOW.Add(S);
Console.WriteLine("The Time After Adding 100Secs : {0}",AfterAdd100Secs);
```
