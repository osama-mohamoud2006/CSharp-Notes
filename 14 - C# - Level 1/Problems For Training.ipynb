### Problem 12
Write a C# program that takes a number as input and displays it four times in a row (separated by spaces), and then four times in the next row with no spaces. Repeat this pattern twice using placeholders (`{0}`).

**Test Data:**

```
Enter a digit: 25
```

**Expected Output:**

```
25 25 25 25
25252525
25 25 25 25
25252525
```

````
byte N = 25;

for(int i=1; i<=4; i++)
{
    if(i%2!=0) // odd 
    {
        Console.WriteLine("{0}  {0}  {0}  {0}  {0}", N, N, N, N);
    }
    else
        Console.WriteLine("{0}{0}{0}{0}{0}", N, N, N, N);
}
````


````
//Check if one is negative and one is positive --> True

bool Check(int n1 , int n2)
{
    return ((n1 > 0 && n2 < 0) || (n1 < 0 && n2 > 0) );
}
Console.WriteLine(Check(10, -10));
Console.WriteLine(Check(-10, -10));
Console.WriteLine(Check(-10, 10));
````

### Problem 19
Write a C# program to compute the sum of two given integers. If the two values are the same, return triple their sum.


````
int Sum(int n1 , int n2)
{
    return (n1 == n2) ? 3*(n1 + n2) :  (n1 + n2);
}
Console.WriteLine(Sum(1, 1));
````

### Problem 20
Write a C# program to get the absolute value of the difference between two given numbers. Return double the absolute value of the difference if the first number is greater than the second number.


````
int Diff(int n1 ,int n2)
{
    return (n1 > n2) ? (2 * Math.Abs(n1 - n2)) : Math.Abs(n1 - n2);
}
Console.WriteLine(Diff(20, 10));
````
