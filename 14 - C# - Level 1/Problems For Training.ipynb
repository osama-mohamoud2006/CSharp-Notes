Write a C# Sharp program to swap two numbers.
Test Data:
Input the First Number : 5
Input the Second Number : 6
Expected Output:
After Swapping :
First Number : 6
Second Number : 5

```
sbyte N = 5;
sbyte N2 = 10;

Console.WriteLine("the Numbers Before Swaping: {0},{1}", N, N2);

var temp = N; // Num1 
N = N2;
N2 = temp;

Console.WriteLine("the Numbers After Swaping: {0},{1}", N, N2);

```
