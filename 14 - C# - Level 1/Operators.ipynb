# Assignment Operators

```
byte x = 5;
// Addition Assignment 
x += 10; // --> x = x+10 

// Subtraction Assignment 
x -= 5; // --> x = x -5

// Multiplication Assignmnet 
x *= 6; // --> x = x * 6

// Divison Assignment 
x /= 1; // --> x = x /1 

// Modulo Assignment 
x %= 2; // ----> x = x %2 
```

 
##### As you know this operation on is done on bit level. 


```
enum Per : byte { Add = 2, Delete = 4, Update = 8 };
// Bitwise OR assignment // 
Per permisson = 0;
permisson |= Per.Add; 
permisson |= Per.Delete; // ------> permission = permission |  Per.Delete
permisson |= Per.Update;

Console.WriteLine(permisson);
```

```
int x = 0;
x |= 10;
x |= 9;
Console.WriteLine(x);
```

```
// Bitwise AND Operation //
using System.Linq.Expressions;

byte A = 10;
byte B = 7;

var x = A & B;
B &= 15; // --> B = B & 15
Console.WriteLine(x);
```

```
// XOR Assignment
x ^= 10;
Console.WriteLine(x);
```

```
// Left Shift Assignment --> - It is NOT a logical operation,It is a bitwise operation It is used to shift the bits of a number to the left by a specific number of steps.
var Bin = 0b1000;
Console.WriteLine((int)Bin);

Bin <<= 0001; // Bin = Bin << 1
Console.WriteLine((int)Bin);

Bin <<= 0010; // Bin = Bin << 2
Console.WriteLine((int)Bin);
```

```
// Rigth Shift Assignment 
var A = 0b1000;
Console.WriteLine((int)A);
A >>= 3; // A = A >> 3 
Console.WriteLine((int)A);
```
