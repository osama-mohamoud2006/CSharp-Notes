##  Bitwise and Bit Shift Operators


###### Bitwise and bit shift operators are used to perform bit manipulation operations.


 
##### As you know this operation on is done on bit level. 


```
enum Per : byte { Add = 2, Delete = 4, Update = 8 };
// Bitwise OR assignment // 
Per permisson = 0;
permisson |= Per.Add; 
permisson |= Per.Delete; // ------> permission = permission |  Per.Delete
permisson |= Per.Update;

Console.WriteLine(permisson);
```

```
// XOR Assignment
x ^= 10;
Console.WriteLine(x);
```

```
// Binary Search --> for sorted array only 
int? BinarySearch(int []lis , int val) // ---> O(log n)
{
    int start = lis[0];
    int end = lis.Length - 1;

    while (start <= end)
    {
        int mid = (start + end) / 2;
        if (!Convert.ToBoolean(lis[mid] ^ val)) return mid;// return the index
        else if (lis[mid] < val) start = mid + 1;
        else if (lis[mid] > val) end = mid - 1;
    }
    return null;
}
```

```
// Left Shift Assignment --> - It is NOT a logical operation,It is a bitwise operation It is used to shift the bits of a number to the left by a specific number of steps.
var Bin = 0b1000;
Console.WriteLine((int)Bin);

Bin <<= 0001; // Bin = Bin << 1
Console.WriteLine((int)Bin);

Bin <<= 0010; // Bin = Bin << 2
Console.WriteLine((int)Bin);
```

```
// Rigth Shift Assignment 
var A = 0b1000;
Console.WriteLine((int)A);
A >>= 3; // A = A >> 3 
Console.WriteLine((int)A);
```

# Relational Operators


###### Relational operators are used to check the relationship between two operands. If the relationship is true the result will be true, otherwise it will result in false. Relational operators are used in decision making and loops.


```
byte x = 100;
Console.WriteLine(x == 2); // Equals to 
Console.WriteLine(x != 2); // isn't equal
Console.WriteLine(x >= 2); // greater than or equal
Console.WriteLine(x <= 2); // less than or equal
```

# Logical Operators


###### Logical operators are used to perform logical operation such as and, or. Logical operators operates on boolean expressions (true and false) and returns boolean values. Logical operators are used in decision making and loops.
###### && : for And

###### || : for Or

###### ! : for NOT (reverse the results)



```
Console.WriteLine(x >= 2 || x > 100); // The first condition is true OR flase = true
Console.WriteLine(x >= 2 && x> 100); // the 2 conditions must be true 
```

```
int c = 1;
Console.WriteLine(true || c++ < 100); // Short-circuit Evaluation --> لو في شرط من اتنين اتحقق بيتجاهل التاني 

Console.WriteLine(false || c++ < 100); // here it will add one to c as the first condition is false so it will check the second condition 
Console.WriteLine(c);

Console.WriteLine(false && c++ > -1); // here the first condition is false no need to check the second condition 
Console.WriteLine(c);

```

# Unary Operators


###### Unlike other operators, the unary operators operates on a single operand.


```
int number = 32,result;
result = +number; // Postive Number 
Console.WriteLine("+number = {0}", +result);

result = -number; // Negative Number 
Console.WriteLine("-number = {0}", -result);

result++; // increment 
Console.WriteLine("result++ = {0}", result);

result--; // decrement 
Console.WriteLine("result-- = {0}", result);

Console.WriteLine(++result); // increment then print 
Console.WriteLine(result++); // print then increment 

Console.WriteLine("!true --> {0}", !true);
```

# Ternary Operator


###### The ternary operator ? : operates on three operands. It is a short hand for `if-then-else` statement. Ternary operator can be used as follows:

###### `variable = Condition? Expression1 : Expression2;`


###### The ternary operator works as follows:

###### If the expression stated by Condition is `true`, the result of Expression1 is assigned to variable.
###### If it is `false`, the result of Expression2 is assigned to variable.
###### used for return values not side affect on variables like ++ or --

```
//(item == 'm') ? Male++ : Female++; // --> ternary operator used to return values , not side effect like ++ (error)
byte myage = 20;
var age = (myage < 30) ? "too old" : "young";
Console.WriteLine(age);
```
