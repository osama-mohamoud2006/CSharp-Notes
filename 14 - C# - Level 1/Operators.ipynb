# Assignment Operators

```
byte x = 5;
// Addition Assignment 
x += 10; // --> x = x+10 

// Subtraction Assignment 
x -= 5; // --> x = x -5

// Multiplication Assignmnet 
x *= 6; // --> x = x * 6

// Divison Assignment 
x /= 1; // --> x = x /1 

// Modulo Assignment 
x %= 2; // ----> x = x %2 
```

 
##### As you know this operation on bin level

```
enum Per : byte { Add = 2, Delete = 4, Update = 8 };

Per permisson = 0;
permisson |= Per.Add; // Bitwise OR assignment 
permisson |= Per.Delete; // ------> permission = permission |  Per.Delete
permisson |= Per.Update;

Console.WriteLine(permisson);
```
