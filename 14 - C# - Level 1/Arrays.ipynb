````
int [] arr = {1,2,3};
Console.WriteLine(arr[0]);
Console.WriteLine(arr[1]);
Console.WriteLine(arr[2]);
Console.WriteLine(arr[3]); // access element bigger than acutal size --> System.IndexOutOfRangeException: Index was outside the bounds of the array.
````

# C# Array Declaration

In C#, here is how we can declare an array.
```csharp
datatype[] arrayName;
```

Here,

- **dataType** - data type like `int`, `string`, `char`, etc
- **arrayName** - it is an identifier

## Let's see an example,
```csharp
int[] age;
```

Here, we have created an array named `age`. It can store elements of `int` type.

But how many elements can it store?

To define the number of elements that an array can hold, we have to allocate memory for the array in C#. For example,
```csharp
// declare an array
int[] age;

// allocate memory for array
age = new int[5];
```

Here, `new int[5]` represents that the array can store 5 elements. We can also say the size/length of the array is 5.

> **Note:** We can also declare and allocate the memory of an array in a single line. For example,
```csharp
int[] age = new int[5];
```


-----------------------------------------------------------------------------------------------------
In C#, arrays are stored on the **heap**.

Even if you declare the variable locally (e.g., `int[] age = new int[5];`), the `new` keyword is the giveaway — it always allocates on the heap. What lives on the stack is just the **reference** (pointer) to that heap memory.

```
Stack                  Heap
┌─────────┐           ┌─────────────────────┐
│   age   │ ───────►  │  [0][0][0][0][0]    │
│ (ref)   │           │  int[5] array data  │
└─────────┘           └─────────────────────┘
```

This is true for all arrays in C#, regardless of the element type — even `int[]`, `bool[]`, etc. store their data on the heap, since arrays are reference types.
