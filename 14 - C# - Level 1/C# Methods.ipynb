# C# Methods

* A method is a block of code which only runs when it is called.

* You can pass data, known as parameters, into a method.

* Methods are used to perform certain actions, and they are also known as functions.

* C# is a fully OOP language , you cannot create a method outside class.

**Create a Method:**
A method is defined with the name of the method, followed by parentheses (). C# provides some pre-defined methods, which you already are familiar with, such as Main(), but you can also create your own methods to perform certain actions:

Example:

 `static void MyMethod() 
  {
    // code to be executed
  }`


You should use **`static`** if you want to call the method without having obejct.

```
/*
Anatomy of a method
[access modifier] [static?] [return type] MethodName([parameters])
{
    // body
}
*/
```

```
class Ex
{
    public void PrintName(string Name) // Non staic method you should create object of this class to use it 
    {
        Console.WriteLine($"HELLO {Name.ToUpper()}" );
    }

}
Ex Obj = new Ex(); // object of class 
Obj.PrintName("osama"); // call method inside class (method member)
```

```
class Ex2
{
    public static void PrintName(string Name) // staic method 
    {
        Console.WriteLine($"HELLO {Name.ToUpper()}");
    }

}
Ex2 Obj = new Ex2(); // object of class 
Ex2.PrintName("osama"); // call static method (classname.method())
```

```
 void DisplayInfo(string name , byte age)
{
    Console.WriteLine($"The Name Is: {name}\nThe Age Is: {age}");
}
DisplayInfo("osama", 20);
```

### Default Parameter Value
You can also use a default parameter value, by using the equals sign (=).

```
void DisplayInfo(string name="Maha", byte age=11) // <---- default parameters starts from right to left 
{
    Console.WriteLine($"The Name Is: {name}\nThe Age Is: {age}");
}
DisplayInfo("Ahmad");
```

### Return Values
void keyword indicates that the method should not return a value.

If you want the method to return a value, you can use a primitive data type (such as int or double) instead of void, and use the return keyword inside the method.

```
int GetNumber()
{
    return 1011;
}
Console.WriteLine(GetNumber());
```

```
class clsPerson
{
    public static string GetFullName(string Fname="maha", string Lname="emad") // default paramters 
    {
        return $"{Fname} {Lname}";
    }
}
var name = clsPerson.GetFullName(); // no compilation error as it has default parameters 
Console.WriteLine(name);
```
