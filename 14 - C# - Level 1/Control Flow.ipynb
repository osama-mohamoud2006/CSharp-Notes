```
/*
if (boolean-expression)
{
	// statements executed if boolean-expression is true
}*/
```

C# if...else (if-then-else) Statement
The if statement in C# may have an optional else statement. The block of code inside the else statement will be executed if the expression is evaluated to `false`.

```
/*
if (boolean-expression)
{
	// statements executed if boolean-expression is true
}
else
{
	// statements executed if boolean-expression is false
}
*/
```

```
bool flag = false;
if (flag)
{
    Console.WriteLine("true");
}
else Console.WriteLine("false");
```

##### C# if...else if (if-then-else if) Statement ,When we have only one condition to test, if-then and if-then-else statement works fine. But what if we have a multiple condition to test and execute one of the many block of code.

```
/*
if (boolean-expression-1)
{
	// statements executed if boolean-expression-1 is true
}
else if (boolean-expression-2)
{
	// statements executed if boolean-expression-2 is true
}
else if (boolean-expression-3)
{
	// statements executed if boolean-expression-3 is true
}
.
.
.
else
{
	// statements executed if all above expressions are false
}

*/
```

```
if (flag)
{
    Console.WriteLine("true");
}
else if (!flag)
{
    Console.WriteLine("false");
}
else
{
    Console.WriteLine("IDK");
}
```

# switch Statement

##### Switch statement can be used to replace the if...else if statement in C#. The advantage of using switch over if...else if statement is the codes will look much cleaner and readable with switch.

##### A problem with the switch statement is, when the matching value is found, it executes all statements after it until the end of switch block.

###### To avoid this, we use `break` statement at the end of each case. The break statement stops the program from executing non-matching statements by terminating the execution of switch statement.

 In C# switch

C# is more flexible.

A switch can work with(The value in each case must be a constant value.):

`int`

`char`

`string`

`enum`

`bool`

some numeric types (`byte`, `short`, `long`)

`float` / `double` (in modern C# through pattern matching)

```
 // switch case in C++ accepts only integral nums , enums
 // here in C# it can accept string , integral nums , enums 
```

```
/*
switch (variable/expression)
{
    case value1:
        // Statements executed if expression(or variable) = value1
        break;
    case value2:
        // Statements executed if expression(or variable) = value1
        break;
    ... ... ... 
    ... ... ... 
    default:
        // Statements executed if no case matches
}
*/
```

```
enum enWeekDays : byte
{
    sat = 1, sun = 2, mon = 3, tue = 4, wed = 5, ths = 6, fri = 7
}
enWeekDays WeekDay = enWeekDays.sat;

// Switch statement with enum 
switch (WeekDay)
{
    case enWeekDays.sat:
        Console.WriteLine(enWeekDays.sat);
        break;

    case enWeekDays.sun:
        Console.WriteLine(enWeekDays.sun);
        break;
    case enWeekDays.mon:
        Console.WriteLine(enWeekDays.mon);
        break;
}
```

```

string[] arr = { "t1", "t2", "t3", "t4", "t5" };

string test = arr[0];

// switch statement with string 
switch (test) // switch in C# can work with string normally :)
{
    // case arr[0].ToString(): //A constant value of type 'string' is expected

    //     Console.WriteLine(0);
    //     break;

    case "t1":
        Console.WriteLine("t1");
        break;

    case "t2":
        Console.WriteLine("t2");
        //break; /// --> (if you didn't use break after case) Control cannot fall through from one case label ('case "t2":') to another
        break;

    case "t3":
    Console.WriteLine("t3");
        break;

    default:
        Console.WriteLine("NAH");
        break;

}
```

```
char c = 'a';
//switch case can work with chars normally 
switch(c)
{
    case 'a':
        {
            Console.WriteLine("a");
            break;
        }
}
```

```
switch (Char.ToUpper(c))
{
    case 'A':
        Console.WriteLine('A');
        break;
}
```

```
char v = 'e';
// switch statement with grouped cases
switch (Char.ToLower(v))
{
    case 'a':
    case 'e':
    case 'o':
    case 'u':
    case 'i':
        Console.WriteLine("Vowel");
        break;

    default:
        Console.WriteLine("Not Vowel");
        break;
}
```

```
// Switch Statement With Bool 
bool b = true;

switch(b)
{
    case true:
        Console.WriteLine("true");
        break;
    case false:
        Console.WriteLine("false");
        break;
}
```

```
// Switch Statement With Float
float x = 0.1f + 0.2f;

switch (x)
{
    case 0.3f:
        Console.WriteLine("Equal");
        break;

    default:
        Console.WriteLine("Not Equal");
        break;
}
```

```
// Switch Statement With Double
double d = 10.2d;
switch(d)
{
    case 10.2d:
        Console.WriteLine("Matched");
        break;
}
```

#  ternary (? :) Operator

We will learn about C# ternary operator and how to use it to control the flow of program.

Ternary operator are a substitute for if...else statement.

Why is it called ternary operator?

This operator takes 3 operand, hence called ternary operator.



The syntax of ternary operator is:

`Condition ? Expression1 : Expression2;`


The ternary operator works as follows:

If the expression stated by Condition is true, the result of Expression1 is returned by the ternary operator.
If it is false, the result of Expression2 is returned.
For example, we can replace the following code:

`if (number % 2 == 0)
{
	isEven = true;
}
else
{
	isEven = false;
}`

```
// Normal if - else 
if (10 % 2 == 0)
{
    return true;
}

else
{
    return false;
}
```

```
// Short Hand If (ternary operator)
Console.WriteLine((10%2==0)? "true" :"false"); // if this condotion is true it will execute the expreesion 1 otherwise it will execute the expression 2 
```

# C# do...while loop

The do and while keyword is used to create a do...while loop. It is similar to a while loop, however there is a major difference between them.

In while loop, the condition is checked before the body is executed. It is the exact opposite in do...while loop, i.e. condition is checked after the body is executed.

This is why, the body of do...while loop will execute at least once irrespective to the test-expression.

The syntax for do...while loop is:

```
byte n = 3;
do
{
    Console.WriteLine("This Body will be executed at least once before looping ");
    Console.WriteLine(n--);
} while (n > 0);
```

```
using System.ComponentModel.DataAnnotations.Schema;

byte n = 3;
do
{
    Console.WriteLine("This Body will be executed at least once before looping");
    Console.WriteLine(n--); // n =2 
} while (n == 3); // the condition must be true to repeat again 
```

# break Statement


C# break Statement
In C#, we use the break statement to terminate the loop.

As we know, loops iterate over a block of code until the test expression is false. However, sometimes we may need to terminate the loop immediately without checking the test expression.

In such cases, the break statement is used. The syntax of break statement is,

break;
Before we learn about break, make sure to learn about

`for loop`
`if...else`
`while loop`

```
byte i = 2;
while (true)
{

    Console.WriteLine(i--);
    if (i == 0) break; // without break it will run infintely 

}
```
