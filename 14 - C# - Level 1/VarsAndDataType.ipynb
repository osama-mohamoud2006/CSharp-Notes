### How To Get The Var Type (Like Py) ?

```
string z = "h";
char c = 'd';
short i = 23;
int x = 100;
Console.WriteLine(z.GetType());
Console.WriteLine(c.GetType());
Console.WriteLine(i.GetType());
Console.WriteLine(x.GetType());

```

### Implicitly Typed Variables

```
var I = 10; // you can never change the data type of implicitly typed var once it is assigned 
//I = "h"; // Cannot implicitly convert type 'string' to 'int'

//var E;   Implicitly-typed variables must be initialized

Console.WriteLine(I);
Console.WriteLine(I.GetType());
```

```
System.Int32 X = 10; // .Net Framework type (int in c#)
Console.WriteLine(X);

System.Int64 LONG = 1939; // .Net Framework type (long in c#)
Console.WriteLine(LONG);


System.Single FL = 10.2f; // .Net Framework type (float in c#)
Console.WriteLine(FL);

System.Double Db = 10.22d; // .Net Framework type (double in c#)
Console.WriteLine(Db);

System.Char C = 'y';
Console.WriteLine(C);

System.Boolean B = true;
Console.WriteLine(B);

byte Num = 255; // range from 0 -> 255
Console.WriteLine(Num);

sbyte Num2 = -128; // singed byte range from -128->127
```
