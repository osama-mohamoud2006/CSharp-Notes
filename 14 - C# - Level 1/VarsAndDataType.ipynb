# Default Values

```
/*
| المكان         | default value     |
| -------------- | ----------------- |
| local variable | ❌ لازم initialize |
| class field    | ✅ بياخد default   |
| struct field   | ✅ بياخد default   |
| array elements | ✅ بياخد default   |
*/

class Test
{
    int i; // data member of the class it will use the default value if you didn't intilize it 

    public void Print()
    {
        Console.WriteLine("{0}", i); // 0
    }


    public void J()
    {
        int i=0; // you should intilize it (local var)
        Console.WriteLine("{0}",i);
    }
}

```

```
int i = default;// C# 7.1 onward
Console.WriteLine(i);

int i2 = default(int); // older c# 
Console.WriteLine(i2);

float F = default;
Console.WriteLine(F);

bool b = default;
Console.WriteLine(b);

char c = default;
Console.WriteLine(c);

//var k = default; //There is no target type for the default literal.
var k = default(float);
```

# Enums

```
enum WeekDays // enum ,if you set a value , it will auto number everything starting from 0 .

{
    Sat = 1,
    Sun = 2,
    Mon = 3
}

//Enum can have any numarical data type byte, sbyte, short, ushort, int, uint, long, or ulong
//but not string
enum enThings : byte
{
    s1 =1, s2=2,s3=3,s4=4
}

// enum FEnum : float //Type byte, sbyte, short, ushort, int, uint, long, or ulong expected
// {
//     f1, f2, f3, f4
// }

// enum SEnum : string //Type byte, sbyte, short, ushort, int, uint, long, or ulong expected
// {
//     f1, f2, f3, f4
// }

enThings T;
T = default; // 0
T = enThings.s1;

Console.WriteLine(T.GetType()); // will print the name of enum as its data type 
Console.WriteLine(T); // will print the enum name 
Console.WriteLine((int)T); // will print the enum value  

// Simple Switch Statement with enum
WeekDays WD= WeekDays.Sun;
switch (WD)
{
    case WeekDays.Sat:
        Console.WriteLine(WeekDays.Sat); // will convert the enum name to string and prints it 
        break;

    case WeekDays.Mon:
        Console.WriteLine(WeekDays.Sat); // will convert the enum name to string and prints it 
        break;

    case WeekDays.Sun:
        Console.WriteLine(WeekDays.Sat); // will convert the enum name to string and prints it
        break;

}
```
