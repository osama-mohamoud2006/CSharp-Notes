```

enum WeekDays // enum ,if you set a value , it will auto number everything starting from 0 .

{
    Sat = 1,
    Sun = 2,
    Mon = 3
}

//Enum can have any numarical data type byte, sbyte, short, ushort, int, uint, long, or ulong
//but not string


// Simple Switch Statement with enum
WeekDays WD = WeekDays.Sun;
switch (WD)
{
    case WeekDays.Sat:
        Console.WriteLine(WeekDays.Sat); // will convert the enum name to string and prints it 
        break;

    case WeekDays.Mon:
        Console.WriteLine(WeekDays.Sat); // will convert the enum name to string and prints it 
        break;

    case WeekDays.Sun:
        Console.WriteLine(WeekDays.Sat); // will convert the enum name to string and prints it
        break;

}


// If you Want To Store The Enum's Name As String 
string EWeek = WeekDays.Sat.ToString();
Console.WriteLine("\nThe Week Name After Converting It To String: " + EWeek);
```

# Nullable Types

```
///int x = null; //Cannot convert null to 'int' because it is a non-nullable value type
Nullable<byte> B = null;
B += 2; // any operation +,*,/ null = null 
Console.WriteLine("{0}", B);

// HasValue -->  true if the current Nullable<T> object has a value; false if the current Nullable<T> object has no value. 

int? x = null; // you can write nullable var like this also 
Console.WriteLine(x.HasValue); // will print false as it is null 


x = (x ?? 0) + 2; // to add 2 to nullable var (null-coalescing operator)

Console.WriteLine("Using GetValueOrDefault(): {0}", x.GetValueOrDefault()); // will print true as it has 2 as a value  

Console.WriteLine("{0},{1}", x.HasValue, x); // will print true as it has 2 as a value  

int? m = 12;
Console.WriteLine(m);
```

# Anonymous Type

```

using System.Data.Common;
using System.Runtime.Intrinsics.X86;
using System.Security.Cryptography;
//you dont specify any type here , automatically will be specified (implicitly typed)
var Person = new { Name = "osama", Id = 164, Age = 20 }; // read only properites , you cann't modify them

Console.WriteLine("The Name From Anonymous var is: " + Person.Name);
Console.WriteLine("The Age Is: {0}", Person.Age);
Console.WriteLine("The Id: {0}", Person.Id);

//Person.Age =22; // anonymous type -- it is read only you cann't modify it 

//An anonymous type's property can include another anonymous type.
var Student = new //anonymous type.
{
    Name = "loaa",
    Age = 22,
    Boy = new { ClassName = "A2", Id = 23 } // another anonymous type in anonymous type
};

Console.WriteLine("\nThe Boy In Class: " + Student.Boy.ClassName);

var temp = Student; // you can store anonymous type in another implicitly typed 
Console.WriteLine("\nThe Boy ID: " + temp.Boy.Id);
```

# Struct 

```
struct stData
{
    // Paramterless Constructor 
    // Constructor will be called only if you use 'new' keyword while you make instance of struct
    public stData() // new feature supported in c# 10 or greater 
    {
        name = "Test,Result From Paramterless Constructor!";
        age = -100;
    }
    public string name;
    public sbyte age;
}
```

```
Console.WriteLine("The Struct Using new keyword:");
// if there is no paramterless constructor then it will use the default values 

stData S = new stData(); // new here isn't related with heap it is normal keyword 
Console.WriteLine(S.age); // will assign the values by default by parameterless constructor 
Console.WriteLine("{0}", S.name);
```

```
// calling the struct without new keyword 
struct Data
{
    public Data()
    {
        name = "";
    }
    public string name;
}

Data f; // struct instance without new keyword + unassigned values+ willn't use the parameterless constructor to intilize values 

void Test()
{
    //Data f =new Data(); // will use the default values if you didn't assign values
    Data f;
    Console.WriteLine(f.name); // ERROR
}
```

```
Console.WriteLine("The Struct without Using new keyword");
stData S1; // created instance of struct without new keyword , its members will n't be intilized(unassigned values) so you should assign values to them

S1.name = "Ahmed"; // you should assign value 
S1.age = 10;

Console.WriteLine("the Name Is {0}", S1.name);
Console.WriteLine("The Age Is: {0}", S1.age);
```

#### Another Example


struct لا يمكن أن يرث من struct آخر أو class.

| الخاصية      | struct    | class  |
| ------------ | --------- | ------ |
| inheritance  | ❌ لا يدعم | ✔ يدعم |
| polymorphism | ❌         | ✔      |


| النوع  | النوع الحقيقي  |
| ------ | -------------- |     
| struct | Value Type     |
| class  | Reference Type |

| النوع  | غالبًا يتخزن في |
| ------ | --------------- |
| struct | Stack           |
| class  | Heap            |



```
struct Person
{
    public Person(string Name, byte Age)  // you can make paramterized constructor in struct 
    {
        name = Name;
        age = Age;
    }
    public Person()
    {
        name = "non";
        age = 0;

    }
    
    public string name;
    public byte age;
    public string address;

    public void GetAddress() // you can make methods normally in stack
    {
        Console.WriteLine("The Address is: {0}", address);
    }

    public void GetAge()
    {
        Console.WriteLine("The Age Is: {0}", age);
    }

    public void GetName()
    {
        Console.WriteLine("The Name Is: {0}", name);
    }
}
```

```

void TestWithoutNew() // used struct without new
{
    // it willn't use the parameterless constructor without new , you should assign values
    Person P1; // i didn't assign values to the struct 
    Console.WriteLine("{0}",P1.name);
    P1.GetAge();
}
TestWithoutNew();
```

```

void TestWithoutNew2() // used struct without new
{
    Person P1; // i assigned values , so no problem 
    P1.name = "kaha";
    P1.age = 24;
    Console.WriteLine("{0}", P1.name);
    P1.GetAge();
}
TestWithoutNew2();
```

```
void UsingNew()
{
    Person p = new Person("lana",12); // you can use paramterized constructor in struct 

   // Person p = new Person(); // if you didn't call the parametrized constructor and it is the only constructor in the struct , it will use the default value it will not affect 

        // if i didn't assign values ,  it would use the default values
    p.GetName();
    p.GetAddress(); // i didn't intilize the address in constructor , so it will use default value
}
UsingNew();
```
