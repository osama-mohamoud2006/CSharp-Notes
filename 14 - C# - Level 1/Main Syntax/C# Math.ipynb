# C# Math
The C# Math class has many methods that allows you to perform mathematical tasks on numbers.

```
Console.WriteLine(Math.Max(100,200)); // Math.Max to return the max num
```

```
Console.WriteLine(Math.BigMul(2, 5)); // multiplication 
```

```
Console.WriteLine(Math.Abs(-100)); // absolute value of number
```

```
Console.WriteLine(Math.Clamp(3,1,20));
```

```
Console.WriteLine(Math.Round(3.44));
```

```
Console.WriteLine(Math.Sqrt(16));
```

```
Console.WriteLine(Math.Ceiling(10.2));
```

```
Console.WriteLine(Math.Truncate(10.323232323));
```

```
Console.WriteLine(Math.PI);
```

```
Console.WriteLine(Math.E);
```

```
Console.WriteLine(Math.Sign(-19));
var x = 10;
if (Math.Sign(x) == -1)
    Console.WriteLine("-");
else if (Math.Sign(x) == 1)
    Console.WriteLine("+");
else Console.WriteLine('0');
```

```
Console.WriteLine(Math.Floor(4.9));
```

```
Console.WriteLine(Math.Pow(2,2));
```
