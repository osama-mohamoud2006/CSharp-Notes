# Hello C# string Methods !

```
// Format --> to store fromat in variable 
// String.Format("formating{0}{1}",arg to pass ) --> return string after formating 
string Str = "Hello string!";
Console.WriteLine(String.Format("This is test \" {0} \" ", Str)); //Composite Formatting
Console.WriteLine("This is test \" {0} \" ", Str);

string Text = $"Hello this is \"{Str}\" modern formating!"; //String Interpolation.
Console.WriteLine(Text);

string Foramting = "This is test \" {0} \" "; //Composite Formatting
Console.WriteLine(Foramting, Str);

string CF = String.Format("Hi {0}(composite formating)", Str);//Composite Formatting
string IF = String.Format($"Hi {Str}");//String Interpolation.
Console.WriteLine(CF);
Console.WriteLine(IF);
```
