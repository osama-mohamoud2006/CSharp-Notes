# Hello C# string Methods !

```
// Format --> to store fromat in variable 
// String.Format("formating{0}{1}",arg to pass ) --> return string after formating 
string Str = "Hello string!";
Console.WriteLine(String.Format("This is test \" {0} \" ", Str)); //Composite Formatting
Console.WriteLine("This is test \" {0} \" ", Str);

string Text = $"Hello this is \"{Str}\" modern formating!"; //String Interpolation.
Console.WriteLine(Text);

string Foramting = "This is test \" {0} \" "; //Composite Formatting
Console.WriteLine(Foramting, Str);

string CF = String.Format("Hi {0}(composite formating)", Str);//Composite Formatting
string IF = String.Format($"Hi {Str}");//String Interpolation.
Console.WriteLine(CF);
Console.WriteLine(IF);
```

```
// split() string
string Record = "Name#//#Balance#//#Id";
string [] SplitedString = Record.Split("#//#");

Console.WriteLine(SplitedString[0]);
Console.WriteLine(SplitedString[1]);
Console.WriteLine(SplitedString[2]);
```

```
// Substring() --> substr() in c++ 
string Record = "Name#//#Balance#//#Id";

string FromIndex = Record.Substring(0, 4); // StartIndex,NumOfCharsToCut 
Console.WriteLine(FromIndex);

string FromIndexToEndString = Record.Substring(7); // StartIndex --> will cut the string from start to end 
Console.WriteLine(FromIndexToEndString);

```

```
// Compare() --> returns num (-1,0,1)
//Lexicographically أبجديًا
string S1 = "cat";
string S2 = "cat";

Console.WriteLine(String.Compare(S1, S2)); // 0 --> The Same Length


string S3 = "Boggy";
string S4 = "Bog";

Console.WriteLine(String.Compare(S3, S4)); // Str1 > Str2 

string s5 = "lol";
string s6 = "LOLOLO";


Console.WriteLine(String.Compare(s5, s6)); // Str1 < Str2 
```

```
string H1 = "HELLO";
string h1 = "hello";

Console.WriteLine(string.Compare(H1, h1,true)); // true --> make the compare case insensitive 
```

```
// CompareTo()
string H1 = "HELLO";
string h1 = "hello";

Console.WriteLine(H1.CompareTo(h1)); // case sensitive 
```

```
// Replace() --> (old value , new value) replace the old value with the new value 
string s = "Hi Loay !";
s = s.Replace("Loay", "Jana");

Console.WriteLine(s);
```

```
// Contains()
string str = "Hello Adel";
Console.WriteLine(str.Contains("Hello")); // return true if(string/char) in the string 
Console.WriteLine(str.Contains('h'));
```

```
// Join() --> concat elements with sperator (sperator , Arr_List_etc)
string[] Arr = { "osama", "ahmed", "maha" };
string res2 = String.Join("-->", Arr);
Console.WriteLine(res2);

int [] Arr2 ={1,2,3,4,5,6,7};
string res = String.Join("#//#", Arr2);
Console.WriteLine(res);

string s = String.Concat("d", "b"); // add 2 strings without sperator 
Console.WriteLine(s);
```

```
// Trim()
string s = "    jk     ";
Console.WriteLine(s);
Console.WriteLine(s.Trim()); // --> removes the ws from left and rigth 
Console.WriteLine(s.TrimStart()); // --> removes the ws from left 
Console.WriteLine(s.TrimEnd());// --> removes the ws from right 
```

```
// EndsWith() --> return true if (string/char) ends with parameter 
string H = "Hello";
Console.WriteLine(H.EndsWith('!'));
Console.WriteLine(H.EndsWith('o'));
Console.WriteLine(H.EndsWith("o"));
```

```
// IndexOf() --> first occurence 
string str = "Lalaaaaa";
Console.WriteLine(str.IndexOf("a", 2)); // start from specific index instead of index 0 
Console.WriteLine(str.IndexOf("a")); // return the first occurence of that char 
Console.WriteLine(str.IndexOf("la")); // return the first occurence of that string 
Console.WriteLine(str.IndexOf('L')); // return the first occurence of that char
```

```
// LastIndexOf() --> the last occurence 
string str = "iam lala";
Console.WriteLine("{0} , {1}", str.LastIndexOf('a'), str.Length - 1);

str = "aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa";
Console.WriteLine("{0} , {1}", str.LastIndexOf('a'), str.Length - 1);
```

```
// Remove()
using System.Globalization;

string s= "Hello World!";
s = s.Remove(s.IndexOf('W')); // remove starting to index to untill the end
Console.WriteLine(s);

s = s.Substring(0, 2); // (start , num of chars) --> it doesn't cut it sclicling string from the total string

Console.WriteLine(s);

s = s.Remove(0, 1);// (start , num of chars) 
Console.WriteLine(s);
```
