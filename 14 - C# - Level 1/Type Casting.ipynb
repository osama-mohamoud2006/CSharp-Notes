# Type Casting

#### Implicit Casting 

```
/*
Implicit Casting (automatically) - converting a smaller type to a larger type size
char -> int -> long -> float -> double
*/
byte b = 10; // --> Max:255
int i = b ; // --> Max:2147483647 , i passed smaller data type value to larger data type 
Console.WriteLine(int.MaxValue);
```

#### Explicit Casting

```
/*
Explicit Casting (manually) - converting a larger type to a smaller size type
double -> float -> long -> int -> char
*/
decimal d = 12.244242m;
byte b = (byte)d; // Explicit Casting , i converted from larger size type to smaller size type 
Console.WriteLine(b);
```

```
string n = "100";
//int num = (int)n; //Cannot convert type 'string' to 'int'
int num = int.Parse(n); // ypu should use int.Prase("string") to convert from string to int 
Console.WriteLine(num.GetType());
```

| **Method** | **Description** |
