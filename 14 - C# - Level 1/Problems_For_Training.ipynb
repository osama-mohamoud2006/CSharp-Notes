26. Sum of First 500 Primes

Write a C# program to compute the sum of the first 500 prime numbers.
Sample Output:
Sum of the first 500 prime numbers:
824693

````
using Microsoft.VisualBasic;

bool IsPrime(int num)
{
    if (1 >= num) return false;
    for (int i = 2; i*i <= num; i++)
    {
        if (num % i == 0) return false;
    }
    return true;
}
int SumOfPrime(int TheSumOfFirst)
{
    int sum = default;
    for (int i = 1; i <= TheSumOfFirst; i++)
    {
        if (IsPrime(i)) sum += i;
    }
    return sum;
}

Console.WriteLine("The Sum Of First 500 Prime Num Is: {0}",SumOfPrime(500));
````
