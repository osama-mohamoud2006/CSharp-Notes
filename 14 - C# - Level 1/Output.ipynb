### WriteLine("") --> For Printing On Console With New Line

```
using System ; 
Console.WriteLine("Hello Iam Osama!"); // WriteLine make new line after printing 
Console.WriteLine("The Age IS: "+20); // Concat
Console.Write("Hi");
Console.Write(" This is single line!");
Console.WriteLine("The Sum Of 2+3= "+(2+3)); // you should use () with concat arithemtic expression with string 
```

### Foramted String(string interpolation)

```
// please note that {0}{1} --> means different arguments to pass 
//  Console.WriteLine("{0}  {0}  {0}  {0}  {0}", N, N, N, N); --> the same argument 
//  Console.WriteLine("{0}  {1}  {2}  {3}  {3}", N1, N2, N3, N4); --> the different argument 

Console.WriteLine("my name is {0} my age is {1}", "osama", 20); // {} must be indexed like {0},{1},..
Console.WriteLine("Hello {0}!", "Lana");
Console.Write("{0}!", "Ahmed");

// you can use $ to write the var within string (String Interpolation)
//$ --> you can make expression,variable in the string without any problem 
string name = "jana";
Console.WriteLine($"\nHi {name}");
```

### Escape Sequence Chars

```
Console.Write("Hi\nOsama\n"); // --> \n New line 
Console.WriteLine("Hi\tOsama"); // ---> \t tab 
Console.WriteLine("Before Backspace\nAfter \bBackspace"); // --> \b Backspace
Console.WriteLine("Hello\'Os\'"); // --> \' Single Quotation 
Console.WriteLine("\"OS\" is imp topic in cs!"); // ---> Double Quotation 
Console.WriteLine("Hi\\Ahmed!"); // --> \\ Back Slach
Console.WriteLine("\a"); // ---> Write Line 
//Console.WriteLine("    c:\source\repos "); // syntax error 
Console.WriteLine(@"    c:\source\repos "); // to use / without needing to make // useful for file paths
```
