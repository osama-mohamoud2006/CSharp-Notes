# C# Exceptions - Try..Catch

When executing C# code, different errors can occur: coding errors made by the programmer, errors due to wrong input, or other unforeseeable things.

When an error occurs, C# will normally stop and generate an error message. The technical term for this is: C# will throw an exception (throw an error).

C# try and catch
The try statement allows you to define a block of code to be tested for errors while it is being executed.

The catch statement allows you to define a block of code to be executed, if an error occurs in the try block.

The try and catch keywords come in pairs:

Syntax:

`try 
{
  //  Block of code to try
}
catch (Exception e)
{
  //  Block of code to handle errors
}`

```
int[] arr = { 1, 2, 3 };

try // exception 
{
    Console.WriteLine(arr[4]);
}
catch (Exception e) // handling 
{
    // Console.WriteLine(e.Message);
    Console.WriteLine(arr[arr.Length-1]);
}
```

```
// Simple Guess Numbe Game

using System;
using System.Collections.Generic;
using System.Linq;
using System.Runtime.InteropServices;
using System.Text;
using System.Threading.Tasks;

namespace ConsoleApp1
{
    class clsGuessGame
    {

        private byte EnterNum()
        {
            byte input;
            bool res = byte.TryParse(Console.ReadLine(), out input);

            while (!res) // the input value isn't valid 
            {
                Console.Write("\aThe Input Isn't Valid\nEnter Valid Input: ");
                res = byte.TryParse(Console.ReadLine(), out input); // Re enter the number 
            }
            return input;
        }

        private byte RandomNumber(byte min, byte max) // to generate random number 
        {
            Random rnd = new Random();
            return (byte)rnd.Next(min, max);
        }

        private bool CheckAnswer(byte ComputerAnswer, byte UserAnswer)
        {
            return (ComputerAnswer == UserAnswer) ? true : false;
        }

        void PrintTheFinalRes(bool res)
        {
            Console.WriteLine();
            if (res) Console.WriteLine("You Won!");
            else Console.WriteLine("You Have Lost!");
        }

        private void RunGame(out bool HasWon)
        {
            HasWon = false;
            byte NumberByComputer = RandomNumber(1, 10);
            Console.Write("\nEnter Number Between 1 -> 10: ");
            byte UserAnswer = EnterNum(); // the Answer From User 

            HasWon = CheckAnswer(UserAnswer: UserAnswer, ComputerAnswer: NumberByComputer);

            PrintTheFinalRes(HasWon); // print the result of match  

            // if HasWon is false then ask user to play again or not 
        }

        private char ReadChar()
        {
            char res;
            bool valid = Char.TryParse(Console.ReadLine(), out res);

            while (!valid || (res.ToString().ToLower() != "y" && res.ToString().ToLower() != "n"))
            {
                Console.Write("\aEnter Valid Char: ");
                valid = Char.TryParse(Console.ReadLine(), out res);
            }
            return res;

        }

        private bool Conform(char ans)
        {
            return (ans.ToString().ToUpper() == "Y") ? true : false;
        }

        public void StartGame()
        {
            Console.Beep();
            Console.WriteLine();

            bool HasWon; // to know if user won or not 
            bool ans = false;
            do
            {
                RunGame(out HasWon); // run the game once 
                                     // if the user has lost then ask him if he wants to try again

                if (!HasWon) // lost 
                {
                    Console.WriteLine();
                    Console.ForegroundColor = ConsoleColor.Red;
                    Console.Write("Do You Want To Try Again? ");
                    ans = Conform(ReadChar()); // 'Y' is true then loop again 
                }
                else
                {
                    ans = false; // don't repeat the game he has won 
                    Console.ForegroundColor = ConsoleColor.Green;
                    Console.WriteLine("Congratulations You Did It!");
                }

            } while (ans);
        }
    }
}
```
