# ReadLine()

```
/*
Reads the next line of characters from the standard input stream.

Returns:
  The next line of characters from the input stream, or null if no more lines are available.
*/
var Input = Console.ReadLine(); // reads anything from user and returns string , 
// please note ReadLine() --> always return string  
Console.WriteLine(Input.GetType());
```

##### input ex :  121 , "oop" , "hi" ,'c' ,12.3 
##### System.String

# Read()

```
/*
Reads the next character from the standard input stream.

Returns:
  The next character from the input stream, or negative one (-1) if there are currently no more characters to be read.
*/
Console.WriteLine("Enter char: ");

var c = Console.Read(); // reads the next character from the std input stream. It returns the ASCII value of the character.
Console.WriteLine();
// please note : even if you entered string like "oop" it will take the first char only 'o' and returns its ascii value 
Console.WriteLine("The Type is: {0}\nThe Value(ASCII) {1}", c.GetType(), c);
```

###### Enter char: 
###### AFFF
###### The Type is: System.Int32
###### The Value(ASCII) 65
