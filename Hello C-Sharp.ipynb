### Input & Output

```
Console.WriteLine("Enter Your Name: ");
string Name = Console.ReadLine(); // to read input from user 
Console.WriteLine("Hello " + Name);

Console.WriteLine("Enter Your Age: ");
//  int Age = Console.ReadLine(); // by default it reads string 
int Age = int.Parse(Console.ReadLine());  // bcz the ReadLine reads string be default , so we use int.Parse() to convert it to int
Console.WriteLine("Your Name is: " + Name + "\nYour Age is: " + Age);

// YourDataType.Parse --> it converts the data type to  YourDataType

string Age = "20"; //string

int IAge = int.Parse(Age); // to int 
double DAge = Double.Parse(Age); // to double 
```

### If Statemnet

```
int x = 10;
var N = 12; // it will determine the data type of value in compile time 

if (N > x)
{
    Console.WriteLine("N IS Bigger Than X");
}
else
{
    Console.WriteLine("N Is Smaller Than X");
}
```

```
// Small Example 
Console.WriteLine("Enter Your Age: "); // print statement on console 
int Age = int.Parse(Console.ReadLine()); // make int input

if (Age == 18)
{
    Console.WriteLine("You Are Adult");
}
else if (Age == 20)
{
    Console.WriteLine("Hello Mr.20");
}
else
{
    Console.WriteLine("Teen Or Very Old");
}
```
